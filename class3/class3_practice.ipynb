### Sales Data Analysis

```python
# You are given weekly sales data of 4 stores for 7 days.
```

```python
sales = np.array([[200, 220, 250, 230, 240, 260, 270],
                  [180, 190, 200, 210, 205, 215, 220],
                  [300, 310, 320, 330, 340, 350, 360],
                  [150, 160, 170, 165, 175, 180, 185]])

#Rows -> stores
#column -> Days
```

```python
# 1️⃣ Total weekly sales of each store

sales.sum(axis=1)
```

```python
# 2️⃣ Average daily sales across all stores

sales.mean(axis = 0)
```

```python
# 3️⃣ Highest sales recorded in the week

sales.max()
```

```python
# 4️⃣ Store with lowest total sales

totals=sales.sum(axis=1)
totals.argmin()
```

```python
# 5️⃣ Increase all sales by 10% using broadcasting

new_sales = sales*1.10
print(new_sales)
```

```python
# 6️⃣ Find sales greater than 300

sales[sales>300]
```

### Temperature Analysis

```python
# You are given temperature (°C) of 6 cities recorded for 5 days.
```

```python
import numpy as np

temp = np.array([[30, 32, 31, 29, 28],
                 [35, 36, 34, 33, 32],
                 [25, 27, 26, 24, 23],
                 [40, 42, 41, 39, 38],
                 [20, 22, 21, 19, 18],
                 [33, 34, 35, 36, 37]])
# Rows -> cities(6)
# columns -> Days(5)
```

```python
#1️⃣ Find average temperature of each city

temp.mean(axis=1)
```

```python
#2️⃣ Find hottest day (day with highest average temperature across cities)

#(Hint: first find column-wise mean, then argmax)

hottest = temp.mean(axis=0)
hottest.argmax()
```

```python
#3️⃣ Find city with lowest minimum temperature

#(Hint: first find row-wise min, then argmin)

lowest_temp =temp.min(axis=0)
lowest_temp.argmin()
```

```python
#4️⃣ Increase temperature of city 3 by 2 degrees (only city 3)

temp[2]+2
```

```python
#5️⃣ Convert all temperatures from Celsius to Fahrenheit

#Formula:

#F = (C × 9/5) + 32

f=temp.copy()
f = (temp * 9/5)+32
f
```

```python
#6️⃣ Find all temperatures above 35°C

temp[temp > 35]
```

### sales dataset
 Monthly sales (in ₹) of 5 stores over 6 months

```python
import numpy as np

# Monthly sales (in ₹) of 5 stores over 6 months
sales = np.array([
    [200, 220, 250, 230, 210, 240],
    [150, 180, 200, 170, 160, 190],
    [300, 320, 310, 330, 340, 350],
    [100, 120, 110, 130, 140, 150],
    [250, 270, 260, 280, 290, 300]
])
```

```python
#1️⃣ Array Basics

#Find the shape and size of the array
sales.shape
sales.size
#Find dtype of the array
print(type(sales))
```

```python
# 2️⃣ Indexing & Slicing

#Print sales of store 3
sales[2]
#Print sales of the first 3 stores for first 4 months
sales[:3,:4]
#Print sales for last 2 stores and last 3 months
sales[-2:,-3:]
```

```python
# 3️⃣ Aggregation

#Find total sales of each store (row-wise sum)
total = sales.sum(axis=1)
total
#Find total sales of each month (column-wise sum)
total_sales = sales.sum(axis=0)
total_sales
#Find maximum sale overall
sales.max()
#Find minimum sale for each store
sales.min()
#Find average sale for each store
sales.mean()
```

```python
# 4️⃣ Broadcasting

#Increase all sales by 10%
new_sales = sales*1.10
print(new_sales)
#Give bonus ₹50 to sales above 300
bonus_sales = sales.copy()
bonus_sales[bonus_sales > 300] +=50
print(bonus_sales)
```

```python
# 5️⃣ Random & Other Functions

#Generate a 3×3 random array of integers between 1–50
arr = np.random.randint(1,51,(3,3))
print(arr)
#Reshape the sales array into 6×5
sales.reshape(6,5)
#Transpose the sales array
sales.transpose()
```

```python
# 6️⃣ Boolean Indexing

#Print all sales above 250
sales[sales>250]
#Print stores that had any month with sales < 150
low_sales = sales[(sales < 150).any(axis=1)]
print(low_sales)
```
