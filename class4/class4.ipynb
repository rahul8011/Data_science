### PANDAS

```python
pandas provide two data stracture:
1. series : only 1D, Homogenous,one column out of a table is series.
    access:
we can access through[]
[] :
    they will work on user defined indexes when we want to get one value using index.
    work on default indexes when used as slicers.
    in slicers,start till end +/-1 based on step size.
    can be used to filter the record
        
.iloc[] -> access through default index.(Dot(.)iloc)
      they will work on default indexes when we want to get one value using index.
      work on default indexes when used as slicers.
      in slicers,start till end +/-1 based on step size.
      CAN'T be used to filter the record
          
.loc[] -> access through user define index(Dot(.)loc)
       they will work on user defined indexes when we want to get one value using index.
       work on user defined indexes when used as slicers.
       in slicers,start till end.
       can be used to filter the record

.ix[] ->
    they will work on user defined indexes when we want to get one value using index.
    work on default indexes when used as slicers.
    in slicers,start till end +/-1 based on step size.
    can be used to filter the record
    
2. DataFrame: only 2D, hetrogenous, coplete table is DataFrame

numpy already installed with pandas

```
