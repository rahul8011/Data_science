```python
### UDF -> User Defined Function
User defined function help to divide a program into smaller, resuable block,
making code more organized,readable,and efficient.They also allow peogrammers 
to reuse logic and share code easily.User Defined Function is created using def keyword.

* Perameter :- The variable written inside the parentheses when defining a Function
    Example: def add(a,b) -> a and b are parameters

* Argument :- The actual value pass to the function when calling it.
    Example: add(5,10) -> 5 and 10 are the arguments

* Return :- The return keyword is use to send the value back from the function to the caller.

* Function call :- A function is executed by writing its name followed by parentheses.
    Example :- function_name()
```

```python
def fu_addTwoNumber(a,b):
    return a + b
```

```python
fu_addTwoNumber(5,10)
```

```python
fu_addTwoNumber(15,10)
```

```python
# Type of Arguments
#1. Positional Arguments :- values are assignes to parameters based on their position
   # order should be correct otherwise output will be wrong
def greet(name,age):
    print(name,age)
    
greet("Rahul",25)
```

```python
#2. keyword Arguments :- values are assigned using parameter names,so order doesn't matter
def greet(age=25, name="Rahul"):
    print("age :", age)
    print("name :", name)
    
greet()
```

```python
#3. Default Argumets:-A parameter can be given a default value when definig a function. 
# If the caller does not provide a value for that parameter, the default value is used
def greet(name,age=25):
    print("Name :-",name)
    print("Age :-",age)
greet("Rahul")
```

```python
def greet(name="Guest"):
    print("Hello",name)
greet()
greet("Rahul")
```

```python
#4. variable-length arguments:- When you want to pass multiple values to a function, but the number of values is not fixed, you can use *args.
# *args collects all positional arguments into a tuple.

def add_numbers(*args):
    print(type(args))
    return sum(args)
    
```

```python
add_numbers(10,20,30)
```

```python
def add_numbers(*nums):
    total = 0
    for i in nums:
        total = total+i
    return total
print(add_numbers(1,2,3,4,5,6))
```

```python
def print_args(*args):
    for i,val in enumerate(args,1):
        print(f"Argument {i}:{val}")
print_args("Rahul",21,"Python")
```

```python
def fn_add3sum(*args):
    Sum = 0
    for var in args:
        if type(var) in (int,float,bool):
            Sum = Sum + var
    return Sum
```

```python
fn_add3sum(10, 20, False, 30, 'a', 40, True, 5.5)
```

```python
# keyword Argument :- **kwargs allow a function to accept any number of keyword arguments, 
#                    which are stored as a dictionary inside the function.
# ** tells python ->  Collect all keyword argument in a dictionary.
# So inside the function, kwargs becomes a dictionary.
# positioning is not important here.  

# Example :-
def student(**kwargs):
    print(kwargs)
student(name = 'Rahul', age = 25, city = 'Delhi')
```

```python
# Accessing the values

def student(**kwargs):
    print("Name :",kwargs["name"])
    print("Age :",kwargs["age"])

student(name="Rahul",age=25)
```

```python
# **kwargs Through loop

def number(**kwargs):
    for key,value in kwargs.items():
        print(key,"=",value)
number(a=10,b=20,c=30)
```

```python
# Using *args and **kwargs together

def together(*args, **kwargs):
    print("Args :",args)
    print("Kwargs :",kwargs)
together(1,2,3,4,5,a=10,b=20,c=30,d=40,e=50)
```

```python
def frn_areaofshapes(shape, **kwargs):
    area = 0
    if shape == 'circle':
        area = 3.14 * kwargs['radius'] ** 2
    elif shape == 'square':
        area = kwargs['side'] * kwargs['side']
    elif shape == 'rectangle':
        area = kwargs['length'] * kwargs['width']
    else:
        area = 'Not valid shape'
    return area
        
```

```python
frn_areaofshapes('square',side=20, length = 20, width=5,radius = 13)
```

### Lambda Function

```python
Properties:
       1. function without the name(anonymous)
       2. can't be reused
       3. can be used to perform only single line operation/s
       4. used when you need a function for a short time and 
        don't want to define it using def.
       5. Lambda functions consume slightly less memory than 
        regular functions because they are anonymous and simpler

Generally they are used along with map(),filter()
to perform the broadcasting
```

```python
# syntax:-

add = lambda n1,n2:n1+n2

result=add(10,15)
print(result)
```

```python
print((lambda n1,n2: n1+n2) (25,30) )
```

```python
Map :- map() is a Python built-in function that applies a given function to each item of an iterable
       and returns a map object containing the results.

      map() is a built-in function used to apply a function to every item of an iterable (like list, tuple, etc.).

      map(function, iterable)
        function â†’ the operation you want to apply
        iterable â†’ list/tuple/string etc.

    map() does NOT return a list directly.
    It returns a map object, so you usually convert it:

map() can work with multiple iterables, but filter() works with only one iterable.
a = [1, 2, 3] # iterable 1
b = [10, 20, 30] # iterable 2

```
