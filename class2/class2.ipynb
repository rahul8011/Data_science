### Data Structure


```python
Properties of python base data strcture:
1. All DS are 1 D, hetrogenous
              1 D :- A one-dimensional (1D) data structure stores data 
                     in a single line or single row.
                  EX:- [10, 20, 30, 40]
              hetrogenous :-A heterogeneous data structure stores different types of data.
                  EX:- [10, "Rahul", 75.5]
                        Integer + String + Float â†’ Heterogeneous.

 
2. Broadcasting is not possible
            Broadcasting:- In Python data structures, broadcasting is not possible means
                           You cannot directly apply a single value to all elements;
                           you must use loops or NumPy.
                               
tuple
list
dict
set

NOTE :- The only difference between list and tuple is list is mutable and tuple is immutable
     Mutable :- we can update the value in side the list on the same memory location 
                that is why they are know as mutable object 
1. How to create them
        tuple : required () or no brackets
        list  : []
(list + Tuple = slicing [start:end:step]
        for reverse we can do it simple[::-1])

        dict  : {key1:value1,key2:value2.......} key value pair seperated by the comma.
            NOTE:- Only immutable object can be use as a key and values can be anything.
                   integer and text use as a key
                   we can not access through index 
        set   : {} also known as keyless dict.
                 always in a sorted manner and there will be not any duplicate value.
                 we can only access the data by iteratable method using loop.
                 there is no any other option to access the data from set
2. How to access element for DS
        tuple : []
        list  : []
        dict  : []
        set   : 
3. How to update them, if possible
4. Methods/Functions of DS
 Methods:- it is a function which is available only for a particular object.
5. How to perform mathematical operation and indexing
```

```python
for var in l1:
    print(var + 200)
```

```python
# add the number 100 to all the elements in the list and update data in the origina list.
for var in range(len(l1)):
    l1[var] = l1[var] + 100
print(l1)
```

```python
# add the number to all the elements in the list and create a new list with updated values.
l1 = [10,20,30,40]
l2 = []
for var in l1:
    l2.append(var + 100)
print(l1)
print(l2)
```

```python
l1 = [10,20,30,40,50]
l2 = []

for var in range(len(l1)):
    l2.append(l1[var] + 100)
print(l1)
print(l2)
```
